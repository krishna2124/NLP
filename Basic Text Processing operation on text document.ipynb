```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.porter import *
from nltk.stem.wordnet import WordNetLemmatizer
```

# Reading json file

```python
import json
```

```python
with open("quora.json", "r") as file:
    data=json.load(file)
ans=[]
for i in data["answers"]:
    ans.append([i['text']])
```

# Text processing

```python
def converttolower(text):
    return text.lower()
for i in range(10):
    ans[i]=[converttolower(ans[i][0])]
ans
```

```python
import string
import re
```

```python
def removeurl(text):
    return re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
for i in range(10):
    ans[i]=[removeurl(ans[i][0])]
ans
```

```python
def removepunctuation(text):
    return text.translate(str.maketrans("","",string.punctuation+"'"+'"'+"’"))
for i in range(10):
    ans[i]=[removepunctuation(ans[i][0])]
ans
```

```python
import nltk
```

```python
def tokenization(text):
    return nltk.word_tokenize(text)
for i in range(10):
    ans[i]=tokenization(ans[i][0])
ans
```

```python
def removestop(text):
    return [word for word in text if not word in stopwords.words()]
for i in range(10):
    ans[i]=removestop(ans[i])
ans
```

```python
porter = PorterStemmer()
lematizer=WordNetLemmatizer()
def stemming(text):
    return [porter.stem(word) for word in text]
for i in range(10):
    ans[i]=stemming(ans[i])
def lematizing(text):
    return [lematizer.lemmatize(word) for word in text] 
for i in range(10):
    ans[i]=lematizing(ans[i])
ans
```

# Creating TDM

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
a=set()
for i in range(10):
    a.update(ans[i])
list(a)
```

```python
import copy
def tdm(lst):
    temp=[]
    for i in a:
        if i in lst:
            temp.append(1)
        else:
            temp.append(0)
    return temp
temp=[]
for i in range(10):
    temp.append(tdm(copy.deepcopy(ans[i])))
print(temp)
```

```python
df = pd.DataFrame(temp,columns =a)
```

```python
df.to_csv('TDM.csv')
```
