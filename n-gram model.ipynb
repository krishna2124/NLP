```python
from nltk import bigrams, trigrams
from collections import Counter, defaultdict
import nltk
```

```python
import json
```

```python
with open("quora.json", "r") as file:
    data=json.load(file)
ans=[]
for i in data["answer"]:
    ans.append([i['answer']])
```

```python
def converttolower(text):
    return text.lower()
for i in range(10):
    ans[i]=[converttolower(ans[i][0])]
ans
```

```python
import re
```

```python
def removeurl(text):
    return re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', text)
for i in range(10):
    ans[i]=[removeurl(ans[i][0])]
ans
```

```python
import string
def removepunctuation(text):
    return text.translate(str.maketrans("","",string.punctuation+"'"+'"'+"’"))
for i in range(10):
    ans[i]=[removepunctuation(ans[i][0])]
ans
```

```python
def tokenization(text):
    return nltk.word_tokenize(text)
for i in range(10):
    ans[i]=tokenization(ans[i][0])
ans
```

# Unigram Model

```python
# Create a placeholder for model
model1 = {w1: 0 for i in ans for w1 in i  }
```

```python
# Count frequency of co-occurance
for sentence in ans:
    for w1 in sentence:
        model1[w1] += 1
print(model1)
```

```python
# Let's transform the counts to probabilities
total_count = float(sum(model1.values()))
model1_prob={}
for w1 in model1:
    model1_prob[w1] =model1[w1]/total_count
model1_prob
```

```python
model={k: v for k, v in sorted(model1_prob.items(), key=lambda item: item[1],reverse=True)}
model
```

# Bigram model

```python
# Create a placeholder for model
model2 = defaultdict(lambda: defaultdict(lambda: 0))
```

```python
# Count frequency of co-occurance
for sentence in ans:
    for w1, w2 in bigrams(sentence, pad_right=True, pad_left=True):
        model2[w1][w2] += 1
```

```python
# Let's transform the counts to probabilities
model2_prob=defaultdict(lambda: defaultdict(lambda: 0))
for w1 in model2:
    total_count = float(sum(model2[w1].values()))
    for w2 in model2[w1]:
        model2_prob[w1][w2] = model2[w1][w2]/total_count
```

```python
dict(model2_prob["implementation"])
```

# Trigram Model

```python
# Create a placeholder for model
model3 = defaultdict(lambda: defaultdict(lambda: 0))
```

```python
# Count frequency of co-occurance
for sentence in ans:
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        model3[(w1, w2)][w3] += 1
```

```python
# Let's transform the counts to probabilities
for w1_w2 in model3:
    total_count = float(sum(model3[w1_w2].values()))
    for w3 in model3[w1_w2]:
        model3[w1_w2][w3] /= total_count
```

```python
dict(model3["is","impossible"])
```

# find probabity of sentence

```python
from copy import deepcopy as dc
```

```python
def findbigramadd1prob(m1,m2,text):
    text=text.split()
    text.insert(0,None)
    prob=1
    m1[None]=1
    for i in range(1,len(text)):
        if text[i] not in m1:
            temp=1/len(m1)
            prob*=temp
        else:
            temp=(m2[text[i-1]][text[i]]+1)/(m1[text[i]]+len(m1))
            prob*=temp
    return prob
```

```python
text="Determine your priorities and focus on them"
findbigramadd1prob(dc(model1),dc(model2),text)
```

```python
text="Determine your priorities and focus them on"
findbigramadd1prob(dc(model1),dc(model2),text)
```

```python
def perplexity(m1,m2,text):
    prob=findbigramadd1prob(dc(m1),dc(m2),text)
    text=text.split()
    N=len(text)
    return ((1/prob)**(1/N))
```

```python
text="Determine your priorities and focus on them"
perplexity(dc(model1),dc(model2),text)
```

```python
text="Determine your priorities and focus them on"
perplexity(dc(model1),dc(model2),text)
```
